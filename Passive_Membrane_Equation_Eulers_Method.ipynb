```python
'''
Sample of Euler's Method
RaghavP
Updated: September 13, 2023
'''

#clearing all variables here because I have had a lot of debugging issues related
#to past variables interfering with the voltage plot I was trying to make
#feel free to remove
%reset

import numpy as np
from matplotlib import pyplot as plt


#set up time. We want a timer that we increment time step by time step
Tmax = 0.5   #500 ms
dt = 0.001   # 1 millisecond time step
t = np.arange(0, Tmax, dt)


#set up biophysical parameters
#remember, for hw assignments don't apply the prefix
#example: Vm = -0.065, NOT Vm=-65
#so we have Cm = capacitance, Vm = voltage, Rm = resistance, I = current
# I use 'Cm' as naming convention as in "Membrane Capacitance"
Cm = 100*10e-12 #picoFarads
Vm = -0.065     #milliVolts
Rm = 4*10e6     #MegaOhms
I = 1*10e-9     #nanoAmperes


#declare+initialize voltage array
V = np.zeros(t.size)

#give initial conditions
#you need an initial condition for solving a differential equation
V[0] = -0.065


'''
euler's method: we use this estimation technique because sometimes there are really
difficult differential equations that are hard to solve and so we just estimate what
the curve looks like 
Passive membrane does have an exact solution (red box, bottom of slide 30), so if you
want, you can plot that and it should exactly overlap what the below for-loop will plot
might be a good exercise to plot the exact solution as a dashed line on this numerical solution below

'''


for i in range(0, t.size-1):
    #v[i+1] = v[i] + dt*function
    V[i+1] = V[i] + dt*(   -(V[i] - Vm)/Rm + I   )/Cm


plt.figure()
plt.title("Passive Membrane Equation")
plt.plot(t, V)
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")
plt.show()

'''
Questions to think about:
1) How is this different from other Artificial Neurons?

2) What makes it more biologically closer?

'''
```

```python
#clearing all variables here because I have had a lot of debugging issues related
#to past variables interfering with the voltage plot I was trying to make
#feel free to remove
%reset

import numpy as np
from matplotlib import pyplot as plt


#set up time
Tmax = 0.5
dt = 0.001
t = np.arange(0, Tmax, dt)


#set up biophysical parameters
#remember, for hw assignments don't apply the prefix
#example: Vm = -0.065, NOT Vm=-65
Cm = 100*10e-12
Vm = -0.065
Rm = 4*10e6

#let's say we want to apply a current for a short period of time.
#in above example, we applied the current the entire time (from start=0ms to 500ms)
#but now we only apply current from start=100ms to 300ms
I = np.zeros(t.size)
I[100:300] = 1*10e-9


#declare+initialize voltage array
V = np.zeros(t.size)

#give initial conditions
V[0] = -0.065


#euler's method: we use this estimation technique because sometimes there are really
#difficult differential equations that can't be solved and so we just estimate what
#the curve looks like 
#Passive membrane does have an exact solution (red box, bottom of slide 30), so if you
#want, you can plot that and it should exactly overlap what the below for-loop will plot
for i in range(0, t.size-1): 
    #v[i+1] = v[i] + dt*function
    V[i+1] = V[i] + dt*(   -(V[i] - Vm)/Rm + I[i]   )/Cm


plt.figure()

plt.subplot(2, 1, 1)
plt.title("Passive Membrane Equation over time")
plt.plot(t, V)
plt.xlabel("time (s)")
plt.ylabel("voltage (V)")

plt.subplot(2, 1, 2)
plt.title("Current over time")
plt.plot(t, I)
plt.xlabel("time (s)")
plt.ylabel("current (Ampere)")

plt.tight_layout()
plt.show()
```
